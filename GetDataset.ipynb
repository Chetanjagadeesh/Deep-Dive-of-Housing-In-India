```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.request import urlretrieve
import ssl
```

```python
base_url = 'https://censusindia.gov.in/2011census/HLO/HL_PCA/'
```

```python
def getURLs(url):
    req = requests.get(url, verify=False)
    soup = BeautifulSoup(req.content, 'html.parser')
    rows = soup.find_all(class_='gridviwalternateitem')
    urls = []
    for row in rows:
        a_tags = row.find_all(href=True)
        if(len(a_tags)>0):
            for a in a_tags:
                urls.append(a['href'])
    return urls
```

```python
# Get the urls for all states
state_urls = getURLs(base_url + 'Houselisting-housing-HLPCA.html')
```

```python
# Get urls for csv files
csv_urls_final = []
for url in state_urls:
    csv_urls = getURLs(base_url + url)
    csv_urls_final += csv_urls
```

```python
ssl._create_default_https_context = ssl._create_unverified_context
for index, url in enumerate(csv_urls_final):
    urlretrieve(base_url+url, f"dataset{index}.xlsx")
```
